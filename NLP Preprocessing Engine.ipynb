# Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love"
In NLP, words are treated based on their exact form. "Love" and "love" are considered different because one is capitalized and the other is not. This can create duplicate meanings. To avoid this, text is usually converted to lowercase so both are treated the same.

## 2. What happens if stopwords are not removed?
If stopwords are not removed, the dataset will contain many common words like "is", "the", and "and" which do not carry important meaning. This increases noise and makes models less efficient.

## 3. When removing stopwords can be harmful
- In sentiment analysis, removing words like "not" can change meaning. Example: "not good" becomes "good".
- In search or question systems, removing words like "what" or "how" may affect understanding of the query.

## 4. Stemming vs Lemmatization
Stemming cuts words into base form without checking meaning. For example, "playing" may become "play" or "pla".
Lemmatization converts words into their correct root form using rules and vocabulary. For example, "playing" becomes "play".

# Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    return tokens
```

# Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens = preprocess_text(s)
    print("\nOriginal:", s)
    print("Tokens:", tokens)
    print("Clean Sentence:", " ".join(tokens))
```

# Task 4: Token Analytics

```python
def analyze(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg = sum(len(w) for w in tokens) / total

    return total, unique, avg

for s in sentences:
    tokens = preprocess_text(s)
    t, u, a = analyze(tokens)

    print("\nSentence:", s)
    print("Total:", t, "Unique:", u, "Avg Length:", round(a,2))
```

# Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []
for s in sentences:
    all_tokens.extend(preprocess_text(s))

counter = Counter(all_tokens)

print("Top 10:", counter.most_common(10))
print("Least 5:", counter.most_common()[-5:])
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    tokens = []
    clean_sentences = []

    for t in text_list:
        tok = preprocess_text(t)
        tokens.extend(tok)
        clean_sentences.append(" ".join(tok))

    return {"tokens": tokens, "clean_sentences": clean_sentences}
```
